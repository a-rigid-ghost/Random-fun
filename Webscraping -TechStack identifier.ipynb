```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
inp='springfest.in'
```

```python
pre='https://builtwith.com/?https%3a%2f%2f'
suff='%2f'
url=pre+inp+suff
```

```python
response=requests.get(url)
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
domains=soup.find_all('div',class_ ="card mt-4 mb-2")
```

```python
total_stack=[]
for domain in domains:
    domain_stack=[]
    stacks=domain.find_all('a',class_ ="text-dark")
    for stack in stacks:
        domain_stack.append(stack.get_text())
    total_stack.append(domain_stack)
```

```python
dom_names=[]
titles=soup.find_all('h6',class_ ="card-title")
len1=len(total_stack)
index=0
for title in titles:
    if(index<len1):
        dom_names.append(title.get_text())
    index=index+1
```

```python
import itertools, pandas
pandas.DataFrame((_ for _ in itertools.zip_longest(*total_stack)), columns=dom_names)
```
